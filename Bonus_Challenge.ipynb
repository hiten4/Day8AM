```python
amount      = float(input('Transaction amount (Rs): '))
category    = input('Category (food/travel/electronics/other): ').lower()
hour        = int(input('Hour of transaction (0-23): '))
daily_spent = float(input('Amount already spent today (Rs): '))
is_vip      = input('Is this a VIP account? (yes/no): ').lower() == 'yes'

# Bonus: ternary operator doubles all limits for VIP
multiplier        = 2 if is_vip else 1
SINGLE_LIMIT      = 50_000  * multiplier
DAILY_LIMIT       = 100_000 * multiplier
FOOD_LIMIT        = 5_000   * multiplier
ELECTRONICS_LIMIT = 30_000  * multiplier

# TODO 1 & 2: BLOCK rules — highest priority, override everything
if amount > SINGLE_LIMIT:
    print(f"BLOCKED: exceeds single transaction limit (Rs.{SINGLE_LIMIT:,.0f})")

elif daily_spent + amount > DAILY_LIMIT:
    print(f"BLOCKED: exceeds daily spending limit (Rs.{DAILY_LIMIT:,.0f})")

# TODO 3: FLAG — unusual hours
elif hour < 6 or hour > 23:
    print(f"FLAGGED: unusual hour ({hour}:00 is outside 06:00–23:00)")

# TODO 4: FLAG — category limits
elif category == 'food' and amount >= FOOD_LIMIT:
    print(f"FLAGGED: food transaction exceeds category limit (Rs.{FOOD_LIMIT:,.0f})")

elif category == 'electronics' and amount >= ELECTRONICS_LIMIT:
    print(f"FLAGGED: electronics exceeds category limit (Rs.{ELECTRONICS_LIMIT:,.0f})")

# TODO 5: All clear
else:
    print("APPROVED")
```
